```python
def add_task(todo_list, task):
    todo_list.append({"task": task, "complete": False})
    print(f"Task '{task}' added successfully!")

def delete_task(todo_list, task_index):
    if 0 <= task_index < len(todo_list):
        deleted_task = todo_list.pop(task_index)
        print(f"Task '{deleted_task['task']}' deleted successfully!")
    else:
        print("Invalid task number.")

def display_tasks(todo_list):
    if not todo_list:
        print("No tasks available.")
    else:
        print("Todo List:")
        for index, task in enumerate(todo_list, start=1):
            status = "Done" if task["complete"] else "Not Done"
            print(f"{index}. {task['task']} - {status}")

def mark_as_complete(todo_list, task_index):
    if 0 <= task_index < len(todo_list):
        todo_list[task_index]["complete"] = True
        print("Task marked as complete!")
    else:
        print("Invalid task number.")

def main():
    todo_list = []

    while True:
        print("\nTodo List Menu:")
        print("1. Add Task")
        print("2. Delete Task")
        print("3. Display Tasks")
        print("4. Mark Task as Complete")
        print("5. Exit")

        try:
            choice = int(input("Enter your choice (1-5): "))
            if choice == 1:
                task = input("Enter the task: ")
                add_task(todo_list, task)
            elif choice == 2:
                display_tasks(todo_list)
                task_index = int(input("Enter the task number to delete: ")) - 1
                delete_task(todo_list, task_index)
            elif choice == 3:
                display_tasks(todo_list)
            elif choice == 4:
                display_tasks(todo_list)
                task_index = int(input("Enter the task number to mark as complete: ")) - 1
                mark_as_complete(todo_list, task_index)
            elif choice == 5:
                print("Exiting the Todo List application. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number.")

if __name__ == "__main__":
    main()
```
